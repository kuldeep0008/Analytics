```python
def ConvertDependents(val):
    if val=="3+":
        return "3"
    else:
        return val
```

```python
Train_loan_df['Dependents']=Train_loan_df['Dependents'].apply(ConvertDependents)
```

```python
Gender=pd.get_dummies(Train_loan_df['Gender'],drop_first=True)
```

```python
Train_loan_df.drop(['Education','Gender','Property_Area','Self_Employed','Loan_Status'], axis=1, inplace=True)
```

```python
Train_loan_df.drop(['Loan_ID','Married'], axis=1, inplace=True)
```

```python
Train_loan_df = pd.concat([Train_loan_df,Gender,Loan_Approved],axis=1)
```

```python
Train_loan_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X= Train_loan_df.drop('loan_approved', axis=1)
y= Train_loan_df['loan_approved']
```

```python
X_train,X_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg= LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
prediction = reg.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test, prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, prediction)
```
